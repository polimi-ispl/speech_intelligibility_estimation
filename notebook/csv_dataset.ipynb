```python
import os
import pandas as pd
```

### Create CSV for original dataset

```python
libri_orig_paths = ['/nas/public/dataset/LibriSpeech/dev-clean']
libri_orig_name = 'dev-clean'

df_col = ['reader', 'book', 'librispeech_folder', 'path', 'transcription_path','SNR', 'noise_type']
dir_df = pd.DataFrame(columns=df_col)

for o in libri_orig_paths:
    
    for s in os.listdir(o):
        
        for b in os.listdir(os.path.join(o,s)):
            files = [ff for ff in os.listdir(os.path.join(o,s,b)) if ff.endswith('.flac')]
            for f in files:
                row = [s, b, libri_orig_name, os.path.join(o, s, b, f),os.path.join(o,s,b,s+'-'+b+'.trans.txt'), float('inf'), None]
                file_info = pd.Series(row, index=df_col)
                dir_df = dir_df.append(file_info, ignore_index = True)
```

```python
dir_df
```

```python
dir_df.to_csv('csv/orig_librispeech_dev-clean.csv', index=False)
```
